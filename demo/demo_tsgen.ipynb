```python
from tsgen.generators import sine, affine, randn
```

```python
start = "1993"
end = "2021"
freq = "1M"

ts = (
    100 +
    sine(start, end, freq, amp=2, n_periods=10) *
    affine(start, end, freq, 0, 10) +
    randn(start, end, freq, std=0.05) *
    affine(start, end, freq, 0, 10) ** 2
)
```

```python
ts.plot()
```

```python
ts.to_frame()
```

```python
start = "2020-10-16T00:00:00Z"
end = "2020-10-16T00:00:42Z"
freq = "100ms"

ts = (
    sine(start, end, freq, amp=2, n_periods=10) +
    affine(start, end, freq, 0, 10) +
    randn(start, end, freq, std=0.5)
)
```

```python
ts.plot()
```

```python
ts.to_frame()
```
