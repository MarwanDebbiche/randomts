```python
from tsgen.generators import cosine, sine, affine, randn
```

```python
start = "2020"
end = "2021"
freq = "1D"

ts = sine(start, end, freq, n_periods=11) + affine(start, end, freq, 0, 5)
```

```python
ts.plot()
```

```python
ts.to_frame()
```

```python
ts = affine(start, end, freq, 0, 5) + randn(start, end, freq, std=0.3) + sine(start, end, freq, n_periods=12)
ts.plot()
```
